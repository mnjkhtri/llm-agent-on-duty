```python
# Agents and Tasks:
from core.agents import Agent, Task

# Tools:
from core.tools import programmer_from_openai, code_runner
```

```python
tools = {
    "programmer_from_openai": programmer_from_openai,
    "code_runner": code_runner,
}
```

```python
problem = """
Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

 

Example 1:

Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped.

Example 2:

Input: height = [4,2,0,3,2,5]
Output: 9

 

Constraints:

    n == height.length
    1 <= n <= 2 * 104
    0 <= height[i] <= 105

"""

agent = Agent("competitive programmer", "to write efficient functions in both time and space complexity")
task = Task(
        f"Write a Python3 function that solves the following problem. \
        The test cases are provided as list of examples. \
        Strictly check if each and every test cases pass using tools at your disposal. There usually are more than one test case.\
        When all test cases are passed, return the function source code. \
        Just the source code. No need to provide use cases yourself. \
        Problem: {problem}",
        outcome = "A function in Python3 named 'main' that passes each and every test cases examples",
        agent = agent,
        tools = tools
    )
print(task.execute())
```
